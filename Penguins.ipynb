```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use(['dark_background'])

import seaborn as sns
sns.set(color_codes=True)
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df.groupby('species').count()
```

```python
df.groupby('species')['species'].count()
```

```python
c = df.groupby('species')['species'].count()
```

```python
plt.pie(c)
plt.show()
```

```python
plt.pie(c, labels=c.index);
```

```python
plt.pie(c, labels=c.index, autopct='%.2f%%');
```

```python
plt.pie(c, labels=c.index, autopct='%.2f%%', explode=[0,1,0]);
```

```python
plt.pie(c, labels=c.index, autopct='%.2f%%', explode=[0,1,0], startangle=180);
```

```python
plt.pie(c, labels=c.index, autopct='%.2f%%', explode=[0,1,0], startangle=180, shadow=True);
```

```python
plt.pie(np.random.randint(0,10,10), wedgeprops=dict(width=0.3))
plt.show()
```

```python
cmap = plt.get_cmap('Accent')
mycolors = cmap(np.arange(10))
```

```python
plt.pie(np.random.randint(0,10,10), wedgeprops=dict(width=0.3), colors=mycolors)
plt.show()
```

```python
plt.pie(c, labels=c.index, wedgeprops=dict(width=0.3), autopct='%0.2f%%')
plt.show()
```

```python
c_i = df.groupby('island')['island'].count()
```

```python
c_i
```

```python
plt.pie(c_i, wedgeprops=dict(width=0.3), labels=c_i.index, autopct='%0.2f%%', labeldistance=1.1);
plt.show()
```
